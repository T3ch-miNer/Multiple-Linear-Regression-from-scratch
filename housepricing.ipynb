```python
data= np.loadtxt('ex1data1.txt',delimiter=',',usecols=range(2))
```

```python
X=data[0:,:1]
y=data[0:,1:]
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
plt.plot(X,y,'+')
plt.show()
```

```python
theta=np.zeros([2,1])
theta.shape

```

```python
m=X.shape[0]
```

```python
x=np.hstack([np.ones([m,1], dtype=np.int32),X])
print(x.shape)
print(theta.shape)
```

```python
iter=1500
alpha=0.01
cost_history=np.ones([iter,1])
for i in range(iter):
    cost=(np.square(x.dot(theta)-y))
    avg_cost=cost.sum()/(2*m)
    
    d=(x.dot(theta)-y)
    d=np.transpose(d)
    d=d.dot(x)

   
    d=alpha*(d/m)
    
    theta=theta-d.reshape(2,1)
    cost_history[i]=avg_cost
print(theta)    
```

```python
plt.plot(X,y,'o',c='g')
Y=x.dot(theta)
plt.plot(X,Y)
plt.show()
theta1=np.linspace(-10,10,1500)
theta2=np.linspace(-10,10,1500)
fig=plt.figure()
ax=fig.add_subplot(111,projection='3d')
ax.set_xlabel('cost_history')
ax.set_ylabel('theta1')
ax.set_zlabel('theta2')
ax.scatter(cost_history,theta1,theta2,c='r',depthshade=True)
plt.show()



```
